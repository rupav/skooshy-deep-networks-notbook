```python
from sklearn.svm import SVR
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate x points and y as sin of those data points.
x = np.sort(10 * np.random.rand(40, 1), axis=0)
y = np.sin(x).ravel()
```

```python
### for every 5th point (40/5 = 8) introduce noice.
y[::5] += 3 * (0.5 - np.random.rand(8))
```

```python
plt.scatter(x, y)
plt.xlabel('data')
plt.ylabel('target')
plt.title('Dummy Data Generated')
plt.show()
```

### Regression Model

```python
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.1)
svr_poly = SVR(kernel='poly', C=1e3, degree=2)
svr_lin = SVR(kernel='linear', C=1e3)
y_rbf = svr_rbf.fit(x, y).predict(x)
y_lin = svr_lin.fit(x, y).predict(x)
y_poly = svr_poly.fit(x, y).predict(x)
```

```python
plt.scatter(x, y, color='orange', label='data')
plt.plot(x, y_rbf, label='RBF model', color='green')
plt.plot(x, y_lin, label='Linear model')
plt.plot(x, y_poly, label='Poly model', color='black')
plt.xlabel('data')
plt.ylabel('target')
plt.title('SVR model with different kernels')
plt.legend()
plt.show()
```

## As can be seen, radial basis function kernel works the best in SVR model, fitting the sinusoidal data with noise.
